```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from PIL import Image
import torch.nn as nn
import pandas as pd
from transformers import SegformerImageProcessor, AutoModelForSemanticSegmentation

class BodyMeasurementProcessor:
    def __init__(self, keypoints_csv, image_dir, mask_dir, output_csv):
        self.keypoints_csv = keypoints_csv
        self.image_dir = image_dir
        self.mask_dir = mask_dir
        self.output_csv = output_csv
        self.df = pd.read_csv(keypoints_csv)

        # Initialize the new DataFrame for output
        self.output_df = pd.DataFrame(columns=['Image_Name', 'Head_Length', 'Hip_Width', 'Hip_Left_Coordinate', 'Hip_Right_Coordinate',
                                               'Waist_Width', 'Waist_Left_Coordinate', 'Waist_Right_Coordinate'])

        # Load the model and processor
        self.processor = SegformerImageProcessor.from_pretrained("mattmdjaga/segformer_b2_clothes")
        self.model = AutoModelForSemanticSegmentation.from_pretrained("mattmdjaga/segformer_b2_clothes")

    def process_segmentation_mask(self, image_path, depth_image_path):
        image = Image.open(image_path).convert("RGB")
        inputs = self.processor(images=image, return_tensors="pt")
        outputs = self.model(**inputs)
        logits = outputs.logits.cpu()
        upsampled_logits = nn.functional.interpolate(logits, size=image.size[::-1], mode="bilinear", align_corners=False)
        pred_seg = upsampled_logits.argmax(dim=1)[0].numpy()

        # Load the original depth image
        depth_image = cv2.imread(depth_image_path, cv2.IMREAD_GRAYSCALE)

        # Check if the depth image was loaded successfully
        if depth_image is None:
            print(f"Error: Unable to load depth image at {depth_image_path}")
            return None  # Return None if the depth image cannot be loaded

        # Resize depth image to match the segmentation mask dimensions
        depth_image_resized = cv2.resize(depth_image, (pred_seg.shape[1], pred_seg.shape[0]))

        # Define classes to exclude (e.g., arms)
        exclude_classes = {"Left-arm": 14, "Right-arm": 15}
        for class_id in exclude_classes.values():
            depth_image_resized[pred_seg == class_id] = 0  # Zero out arm areas in the depth mask

        return depth_image_resized
    def get_keypoint(self, result, keypoint_id):
        keypoint_column = f'Keypoint{keypoint_id}'
        if keypoint_column in result.columns:
            keypoint_data = result[keypoint_column].values[0]
            if not pd.isna(keypoint_data):
                return np.array(eval(keypoint_data))  # Assuming the data is stored as a string representation of a list
        return None

    def find_nearest_black_pixels(self, image_mask, y_coord, center_x):
        left_coord = right_coord = None
        for x in range(center_x, -1, -1):
            if image_mask[y_coord, x] == 0:
                left_coord = (x, y_coord)
                break
        for x in range(center_x, image_mask.shape[1]):
            if image_mask[y_coord, x] == 0:
                right_coord = (x, y_coord)
                break
        return left_coord, right_coord

    def display_results(self, image_name, result):
        image_path = f"{self.image_dir}/{image_name}"
        depth_image_path = f"{self.mask_dir}/{image_name}"

        # Try to process the segmentation mask
        img_mask = self.process_segmentation_mask(image_path, depth_image_path)
        if img_mask is None:
            print(f"Skipping image: {image_name} due to missing or invalid depth image.")
            return

        keypoints = {
            "nose": self.get_keypoint(result, 0),
            "left_ear": self.get_keypoint(result, 3),
            "right_ear": self.get_keypoint(result, 4),
            "left_shoulder": self.get_keypoint(result, 6),
            "right_shoulder": self.get_keypoint(result, 7),
            "left_hip": self.get_keypoint(result, 12),
            "right_hip": self.get_keypoint(result, 13)
        }

        head_length = None
        if keypoints['left_ear'] is not None and keypoints['right_ear'] is not None:
            head_length = np.linalg.norm(keypoints['left_ear'] - keypoints['right_ear'])

        hip_width = waist_width = None
        hip_left_coord = hip_right_coord = waist_left_coord = waist_right_coord = None

        if keypoints['left_hip'] is not None and keypoints['right_hip'] is not None:
            y_hip = int((keypoints['left_hip'][1] + keypoints['right_hip'][1]) / 2)
            hip_center_x = int((keypoints['left_hip'][0] + keypoints['right_hip'][0]) / 2)

            # Ensure y_hip is within image bounds
            y_hip = min(y_hip, img_mask.shape[0] - 1)

            hip_left_coord, hip_right_coord = self.find_nearest_black_pixels(img_mask, y_hip, hip_center_x)
            hip_width = np.linalg.norm(np.array(hip_left_coord) - np.array(hip_right_coord)) if hip_left_coord and hip_right_coord else 0

            if keypoints['nose'] is not None:
                nose_y = keypoints['nose'][1]
            else:
                nose_y = (keypoints['left_shoulder'][1] + keypoints['right_shoulder'][1]) / 2  # Fallback to shoulder height

            center_shoulder = (keypoints['left_shoulder'] + keypoints['right_shoulder']) / 2
            y_waist = int(y_hip * 7 / 9 + (nose_y + center_shoulder[1]) / 16)

            # Ensure y_waist is within image bounds
            y_waist = min(y_waist, img_mask.shape[0] - 1)

            waist_center_x = hip_center_x
            waist_left_coord, waist_right_coord = self.find_nearest_black_pixels(img_mask, y_waist, waist_center_x)
            waist_width = np.linalg.norm(np.array(waist_left_coord) - np.array(waist_right_coord)) if waist_left_coord and waist_right_coord else 0

            print(f"Image Name: {image_name}")
            print(f"Hip Width: {hip_width} pixels")
            print(f"Waist Width: {waist_width} pixels")
            print(f"Hip Coordinate (y_hip): {y_hip}, Hip Center (x): {hip_center_x}")
            print(f"Waist Coordinate (y_waist): {y_waist}")
            print(f"Hip Left Coordinate: {hip_left_coord}, Hip Right Coordinate: {hip_right_coord}")
            print(f"Waist Left Coordinate: {waist_left_coord}, Waist Right Coordinate: {waist_right_coord}")

            # Visualization
            color_image = cv2.cvtColor(img_mask, cv2.COLOR_GRAY2BGR)
            if head_length is not None:
                cv2.line(color_image, tuple(keypoints['left_ear'].astype(int)), tuple(keypoints['right_ear'].astype(int)), (255, 255, 0), 2)  # Yellow line for head width
            if hip_left_coord and hip_right_coord:
                cv2.line(color_image, hip_left_coord, hip_right_coord, (0, 255, 0), 2)  # Green line for hip width
            if waist_left_coord and waist_right_coord:
                cv2.line(color_image, waist_left_coord, waist_right_coord, (0, 0, 255), 2)  # Red line for waist width
            cv2.circle(color_image, (hip_center_x, y_hip), 5, (0, 0, 255), -1)  # Blue dot for hip center coordinate
            cv2.circle(color_image, (waist_center_x, y_waist), 5, (255, 0, 0), -1)  # Red dot for waist coordinate

            plt.figure(figsize=(12, 6))
            plt.subplot(1, 2, 1)
            plt.title("Depth Image with Arms Removed")
            plt.imshow(img_mask, cmap='gray')
            plt.subplot(1, 2, 2)
            plt.title("Measurements on Depth Image")
            plt.imshow(cv2.cvtColor(color_image, cv2.COLOR_BGR2RGB))
            plt.show()

            # Save measurements to output DataFrame
            self.output_df = self.output_df.append({
                'Image_Name': image_name,
                'Head_Length': head_length,
                'Hip_Width': hip_width,
                'Hip_Left_Coordinate': hip_left_coord,
                'Hip_Right_Coordinate': hip_right_coord,
                'Waist_Width': waist_width,
                'Waist_Left_Coordinate': waist_left_coord,
                'Waist_Right_Coordinate': waist_right_coord
            }, ignore_index=True)

        else:
            print(f"Hip keypoints not found for image: {image_name}")


    def process_all_images(self):
        for image_name in self.df['Image_Name']:
            result = self.df[self.df['Image_Name'] == image_name]
            self.display_results(image_name, result)
        
        # Save the output DataFrame to CSV
        self.output_df.to_csv(self.output_csv, index=False)

image_dir = r"C:\Users\USER\Desktop\all\crop_regroup\Fullkorean"
mask_dir = r"C:\Users\USER\Desktop\all\crop_regroup\Fullkorean_mask"
output_csv = r'C:\Users\USER\Desktop\all\crop_regroup\fullkorean_cordinate.csv'
processor = BodyMeasurementProcessor(r"C:\Users\USER\Desktop\all\crop_regroup\filtered_fullkorean.csv", image_dir, mask_dir, output_csv)
processor.process_all_images()

```
