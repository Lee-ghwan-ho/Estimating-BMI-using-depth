```python
import cv2
import numpy as np

def calculate_white_area(image_path):
    # Read the image
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    # Threshold the image to get only white pixels
    _, thresholded = cv2.threshold(image, 240, 255, cv2.THRESH_BINARY)
    
    # Calculate the number of white pixels
    white_area = np.sum(thresholded == 255)
    
    # Calculate the total number of pixels
    total_area = image.shape[0] * image.shape[1]
    
    # Calculate the ratio of white area to the total area
    white_area_ratio = white_area / total_area
    
    return white_area, white_area_ratio

# Path to the image
image_path = r"C:\Users\USER\Desktop\removehead_purple remove\korean\result\M0120201_M_26_1779_784_depth\1.png"

# Calculate the white area and its ratio
white_area, white_area_ratio = calculate_white_area(image_path)

print(f"White area: {white_area} pixels")
print(f"White area ratio: {white_area_ratio:.2f}")

```

```python
import os
import cv2
import numpy as np
from shutil import copyfile

#  
base_path = r"C:\Users\USER\Desktop\removehead_purple remove\korean\result"
segmentation_path = r"C:\Users\USER\Desktop\removehead_purple remove\korean\201301401_mask"

#    
os.makedirs(segmentation_path, exist_ok=True)

#      ( )
min_white_ratio = 0.10  # 8%
max_white_ratio = 0.15  # 15%

#   
for folder_name in os.listdir(base_path):
    folder_path = os.path.join(base_path, folder_name)
    
    #  
    if os.path.isdir(folder_path):
        # png  
        for file_name in os.listdir(folder_path):
            if file_name.endswith('.png'):
                file_path = os.path.join(folder_path, file_name)
                
                #  
                img = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
                
                #    
                total_pixels = img.shape[0] * img.shape[1]
                
                #   
                white_pixels = np.sum(img == 255)
                
                #    
                white_ratio = white_pixels / total_pixels
                
                #        
                if min_white_ratio <= white_ratio <= max_white_ratio:
                    destination_file_name = folder_name.replace('_depth', '') + ".png"
                    destination_file_path = os.path.join(segmentation_path, destination_file_name)
                    copyfile(file_path, destination_file_path)
                    print(f"Saved {destination_file_name} to {segmentation_path} (White area ratio: {white_ratio:.4f})")
```

```python
import os
import cv2
import numpy as np
from shutil import copyfile

#  
base_path = r"C:\Users\USER\Desktop\new\result"
segmentation_path = r"C:\Users\USER\Desktop\bmi\new_mask"

#    
os.makedirs(segmentation_path, exist_ok=True)

#   
for folder_name in os.listdir(base_path):
    folder_path = os.path.join(base_path, folder_name)
    
    #  
    if os.path.isdir(folder_path):
        max_white_area = 0
        max_white_file = None
        
        # png  
        for file_name in os.listdir(folder_path):
            if file_name.endswith('.png'):
                file_path = os.path.join(folder_path, file_name)
                
                #  
                img = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
                
                #   
                white_area = np.sum(img == 255)
                
                #      
                if white_area > max_white_area:
                    max_white_area = white_area
                    max_white_file = file_path
        
        #        
        if max_white_file:
            destination_file_name = folder_name.replace('_depth', '') + ".png"
            destination_file_path = os.path.join(segmentation_path, destination_file_name)
            copyfile(max_white_file, destination_file_path)
            print(f"Saved {destination_file_name} to {segmentation_path}")
```
