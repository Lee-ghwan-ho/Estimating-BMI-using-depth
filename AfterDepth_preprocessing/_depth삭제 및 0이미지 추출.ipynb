```python
import os

# 경로 설정
directory = r"C:\\Users\\USER\\Desktop\\800\\depth"

# 디렉토리 내 파일 이름 변경
for filename in os.listdir(directory):
    # 파일이 실제로 이미지 파일인지 확인
    if filename.endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
        # 파일 이름에서 '_depth' 제거
        new_filename = filename.replace('_depth', '')

        # 파일 경로 생성
        old_file_path = os.path.join(directory, filename)
        new_file_path = os.path.join(directory, new_filename)

        # 파일 이름 변경
        os.rename(old_file_path, new_file_path)
        print(f"Renamed: {filename} -> {new_filename}")

print("파일 이름 변경 완료.")
```

```python
import os
import shutil

# 원본 디렉토리와 대상 디렉토리 설정
source_directory = r"C:\\Users\\USER\\Desktop\\GWANHO\\segment-anything-main\\notebooks\\result"
target_directory = r"C:\\Users\\USER\\Desktop\\800\\mask"

# 대상 디렉토리가 없으면 생성
if not os.path.exists(target_directory):
    os.makedirs(target_directory)

# 원본 디렉토리의 폴더를 순회
for folder_name in os.listdir(source_directory):
    folder_path = os.path.join(source_directory, folder_name)

    # 폴더인지 확인
    if os.path.isdir(folder_path):
        # 폴더 내에 0 이미지가 있는지 확인
        image_path = os.path.join(folder_path, "0.png")
        if os.path.exists(image_path):
            # 이미지를 대상 디렉토리에 복사하고 이름 변경
            target_path = os.path.join(target_directory, f"{folder_name}.png")
            shutil.copy(image_path, target_path)
            print(f"Copied: {image_path} -> {target_path}")

print("모든 이미지 복사 완료.")
```
